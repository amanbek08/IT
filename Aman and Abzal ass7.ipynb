```python
def hamming1(block1):
    print(block1)
    # create list out of inputted block
    block = []
    for a in block1:
        block.append(int(a))
    # counter of errors
    c = 0
    # check for p1
    p1 = (block[3] + block[5] + block[7]) % 2
    print(p1)
    if p1 != block[1]:
        c = c + 1
        print("Incorrect")
    else:
        print("Correct")
    # check for p2    
    p2 = (block[3] + block[6] + block[7]) % 2
    print(p2)
    if p2 != block[2]:
        c = c + 1
        print("Incorrect")
    else:
        print("Correct")
    # check for p4    
    p4 = (block[5] + block[6] + block[7]) % 2
    print(p4)
    if p4 != block[4]:
        c =c + 1
        print("Incorrect")
    else:
        print("Correct")
    # check for p0    
    p0 = 0
    for i in range(len(block)):
        p0 = p0 + block[i]
    p0 %= 2
    print(p0)
    
    if p0 != block[0]:
        c = c + 1
        print("Incorrect")
    else:
        print("Correct")
    
    # detect location of error
    if c != 0:
        row = 0
        column = 0
        if p1 == block[1] and p2 == block[2]:
            column = 0
        if p1 != block[1] and p2 == block[2]:
            column = 1
        if p1 == block[1] and p2 != block[2]:
            column = 2
        if p1 != block[1] and p2 != block[2]:
            column = 3
        if p4 != block[4]:
            row = 1
        print("Error at position:")
        print(row * 4 + column)
        # flip the errored bit
        if block[row * 4 + column] == 0:
            block[row * 4 + column] = 1
        else:
            block[row * 4 + column] = 0
        print("Corrected bitstring: " + "".join(str(y) for y in block))
    output = ""
    output += str(block[3])
    output += str(block[5])
    output += str(block[6])
    output += str(block[7])
    # return the initial block
    return output
```

```python
def hamming2(block2):
    print(block2)
    # create list out of inputted block
    block = []
    for a in block2:
        block.append(int(a))
    # counter of errors
    c = 0
    # check for p1
    p1 = (block[5] + block[9] + block[13] + block[3] + block[7] + block[11] + block[15]) % 2
    print(p1)
    if p1 != block[1]:
        c = c + 1
        print("Incorrect")
    else:
        print("Correct")
    # check for p2
    p2 = (block[6] + block[10] + block[14] + block[3] + block[7] + block[11] + block[15]) % 2
    print(p2)
    if p2 != block[2]:
        c = c + 1
        print("Incorrect")
    else:
        print("Correct")
    # check for p4
    p4 = (block[5] + block[6] + block[7] + block[12] + block[13] + block[14] + block[15]) % 2
    print(p4)
    if p4 != block[4]:
        c = c + 1
        print("Incorrect")
    else:
        print("Correct")
    # check for p8
    p8 = (block[9] + block[10] + block[11] + block[12] + block[13] + block[14] + block[15]) % 2
    print(p8)
    if p8 != block[8]:
        c = c + 1
        print("Incorrect")
    else:
        print("Correct")
    # check for p0
    p0 = 0
    for i in range(len(block)):
        p0 += block[i]
    p0 %= 2
    print(p0)
    if p0 != block[0]:
        c = c + 1
        print("Incorrect")
    else:
        print("Correct")
    # Detect location of errored bit
    if c != 0:
        row = 0
        column = 0
        if p1 == block[1] and p2 == block[2]:
            column = 0
        if p1 != block[1] and p2 == block[2]:
            column = 1
        if p1 == block[1] and p2 != block[2]:
            column = 2
        if p1 != block[1] and p2 != block[2]:
            column = 3
        if p4 == block[4] and p8 == block[8]:
            row = 0
        if p4 != block[4] and p8 == block[8]:
            row = 1
        if p4 == block[4] and p8 != block[8]:
            row = 2
        if p4 != block[4] and p8 != block[8]:
            row = 3
        print("Error at position: " + str(row * 4 + column))
        # flip the errored bit
        if block[row * 4 + column] == 0:
            block[row * 4 + column] = 1
        else:
            block[row * 4 + column] = 0
        # Output
        print("Corrected block: " + "".join(str(i) for i in block))
    output = ""
    output += str(block[3])
    output += str(block[5])
    output += str(block[6])
    output += str(block[7])
    output += str(block[9])
    output += str(block[10])
    output += str(block[11])
    output += str(block[12])
    output += str(block[13])
    output += str(block[14])
    output += str(block[15])
    return output
```
