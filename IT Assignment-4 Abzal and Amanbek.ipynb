```python
# coded string
code = '100110111101011100'
# variable for decoded string
decode = ''
# while loop which iterates the for loop till the string is completely encoded
while code != '':
    # for loop for encoding
    for i in my_list:
        # if we find the matching bit code at the start of the string then
        if code.find(my_list[i]) == 0:
            # we add that coded character to the decoded string variable
            decode = decode + i
            # then we remove this coded bit
            code = code[len(my_list[i]):]
# print the result
decode
```
