```python
# Read the txt file and store it in variable 's'
with open ("file.txt", "r") as myfile:
    s=myfile.read()
# Displaying our input
print(s)
print('')
# Calculating the number of all characters in an input
l = len(s)
# Stored each unique characters of an input in a list
a = list(set(s))
# Calculated the number of unique characters
n = len(a)
# Started loop that starts from 0 and ends with number of unique characters
for i in range (n):
    # Printed the unique character and it's probability by dividing the number of occurance which 
    # we calculated by using s.count by number of all characters in an input
    print(a[i],' - ',s.count(a[i])/l)
```
